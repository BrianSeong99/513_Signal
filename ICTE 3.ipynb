```python
import numpy as np
import matplotlib.pyplot as plt

segment = 10
left_bound = -6
right_bound = 6
t = np.arange(left_bound, right_bound, 1/segment)

tri = np.empty([0])
for i in np.arange(left_bound, right_bound, 1/segment):
    if i < -1:
        y = 0
    elif i >=-1 and i < 0:
        y = 1 + i
    elif i == 0:
        y = 1
    elif i > 0 and i <= 1:
        y = 1 - i
    else:
        y = 0
    tri = np.append(tri, [y])

unit = np.zeros((right_bound - left_bound) * segment)
current_num = left_bound
for i in range(np.size(unit)):
    if current_num >= -5 and current_num <= 0:
        unit[i] = 4
    else:
        unit[i] = 0
    current_num = current_num + 1/segment

y = np.convolve(unit, tri, 'same')

plt.figure(0)
plt.plot(t, unit, label='Signal Unit')
plt.plot(t, tri, label='Signal Triangle')
plt.legend()

plt.figure(1)
t2 = np.arange(left_bound, right_bound, (right_bound - left_bound)/np.size(y))
plt.plot(t, y, label='Convolution of unit and tri signal')

plt.show()
```
